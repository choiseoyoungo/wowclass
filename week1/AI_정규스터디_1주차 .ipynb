```python
# 빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
# 빙어 데이터 시각화
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 두 데이터 합치기
length = bream_length+smelt_length
weight = bream_weight+smelt_weight

print(length)
print(weight)
```

```python
# 데이터 입력 구조 바꾸기
fish_data = [[l, w] for l, w in zip(length, weight)]

print(fish_data)
```

```python
# 정답지 만들기
fish_target = [1]*35 + [0]*14
print(fish_target)
```

```python
# knn 분류기
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

```python
# 모델 훈련
kn.fit(fish_data, fish_target)
```

```python
# 모델 성능
kn.score(fish_data, fish_target)
```

```python
new_fish = [ [30, 600],[12, 10],[25, 150]]
```

```python
# 2. 기존 도미, 빙어 데이터 + new_fish 시각화
import matplotlib.pyplot as plt

# 기존 데이터 그래프
plt.scatter(bream_length, bream_weight, label='Bream')
plt.scatter(smelt_length, smelt_weight, label='Smelt')

# 새로운 데이터 (다른 마커/색)
plt.scatter([f[0] for f in new_fish],
            [f[1] for f in new_fish],
            marker='^', color='red', label='New Fish')

plt.xlabel('length')
plt.ylabel('weight')
plt.legend()
plt.show()
```

```python
# 3. 새로운 데이터 예측
pred = kn.predict(new_fish)

print("예측 결과:", pred)

# 결과 해석
for i, result in enumerate(pred):
    if result == 1:
        print(f"{new_fish[i]} → 도미")
    else:
        print(f"{new_fish[i]} → 빙어")
```
